```python
import pandas as pd
import numpy as np

from binance.client import Client
```

```python
client = Client()
```

```python
def extract_data_1_min_interval(symbol, start_date, end_date):
    frame = pd.DataFrame(client.get_historical_klines(symbol, '1m', start_date, end_date))
    frame = frame[[0, 4]]
    frame.columns = ['Time', symbol]
    frame = frame.set_index('Time')
    frame.index = pd.to_datetime(frame.index, unit='ms')
    frame = frame.astype(float)

    #frame.to_csv(f'{symbol}_{start_date}_{end_date}_1_MIN_INVERVAL', index=True)
    display(frame)
```

```python
symbols = ['BTCEUR', 'ETHEUR', 'ADAEUR', 'SOLEUR']
start_date = '2022-03-29'
end_date = '2022-04-01'

for symbol in symbols:
    extract_data_1_min_interval(symbol, start_date, end_date)
    print(f'{symbol} FINISHED')
```
